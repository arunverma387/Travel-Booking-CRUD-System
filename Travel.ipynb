```python
import uuid
import sys

class TripBox:
    def __init__(self, nme, goTo, dtz, pplCnt, tpK):
        self.uid = str(uuid.uuid4())[:8].upper()
        self.nm = nme
        self.dst = goTo
        self.dt = dtz
        try:
            self.ppl = int(pplCnt)
        except:
            self.ppl = 1
        self.tp = tpK
        self.st = "Confirmed"

    def __str__(self):
        r = ""
        r += "\nID: " + self.uid + "\n"
        r += "Name: " + self.nm + "\n"
        r += "Place: " + self.dst + "\n"
        r += "When: " + self.dt + "\n"
        r += "Kind: " + self.tp + "\n"
        r += "Num: " + str(self.ppl) + "\n"
        r += "State: " + self.st + "\n"
        return r

allData = []

def findOne(xid):
    xid = xid.strip().upper()
    for z in allData:
        if z.uid == xid:
            return z
    return None

def askIt(t, cc=str):
    while True:
        aa = input(t).strip()
        if aa == "":
            print("Enter something.")
            continue
        try:
            return cc(aa)
        except:
            print("Try again.")

def makeB():
    print("\n-- new entry --")
    a1 = askIt("name: ")
    a2 = askIt("where to: ")
    a3 = askIt("date: ")
    a4 = askIt("how many ppl: ", int)
    a5 = askIt("type: ")
    bx = TripBox(a1, a2, a3, a4, a5)
    allData.append(bx)
    print("added with id:", bx.uid)

def allshow():
    if len(allData) == 0:
        print("\nno data yet")
        return
    print("\n------ LIST ------")
    for p in allData:
        print(p)
    print("total:", len(allData))

def lookfor():
    gg = input("id pls: ").strip()
    f = findOne(gg)
    if f:
        print("\nfound:\n", f)
    else:
        print("not found:", gg.upper())

def change():
    kk = input("id to edit: ").strip()
    f = findOne(kk)
    if not f:
        print("no record")
        return

    print("\nediting:", f.uid)

    nd = input("new place (enter skip to ignore): ").strip()
    if nd != "":
        f.dst = nd

    stt = input("new status (optional): ").strip()
    if stt != "":
        f.st = stt

    print("updated.\n", f)

def drop():
    xx1 = input("id delete: ").strip()
    f = findOne(xx1)
    if not f:
        print("no such id")
        return
    yy = input("sure? y/n: ").lower().strip()
    if yy == "y":
        allData.remove(f)
        print("deleted")
    else:
        print("cancelled")

def menu():
    print("\n== menu ==")
    print("1 add")
    print("2 list")
    print("3 search")
    print("4 edit")
    print("5 remove")
    print("6 exit")

def main():
    allData.append(TripBox("Alice", "London", "2026-06-15", 2, "Flight"))
    allData.append(TripBox("Bob", "Bali", "2026-01-20", 4, "Hotel"))
    print("loaded", len(allData), "items")

    while True:
        menu()
        op_cho = input("pick: ").strip()

        if op_cho == "1":
            makeB()
        elif op_cho == "2":
            allshow()
        elif op_cho == "3":
            lookfor()
        elif op_cho == "4":
            change()
        elif op_cho == "5":
            drop()
        elif op_cho == "6":
            print("bye")
            sys.exit(0)
        else:
            print("wrong opt")

if __name__ == "__main__":
    main()
```
