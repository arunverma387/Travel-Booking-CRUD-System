```python
import uuid
import sys
class Booking:
    def __init__(self, customer_name, destination, travel_date, num_travelers, booking_type):
        self.booking_id = str(uuid.uuid4())[:8].upper()  
        self.customer_name = customer_name
        self.destination = destination
        self.travel_date = travel_date
        self.num_travelers = int(num_travelers) if str(num_travelers).isdigit() else 1 
        self.booking_type = booking_type 
        self.status = "Confirmed"
    def __str__(self):
        return (f"\n--- Booking ID: {self.booking_id} ---\n"
                f"  Customer: {self.customer_name}\n"
                f"  Destination: {self.destination}\n"
                f"  Date: {self.travel_date}\n"
                f"  Type: {self.booking_type}\n"
                f"  Travelers: {self.num_travelers}\n"
                f"  Status: {self.status}\n"
                "--------------------------------")
DB = []
def F(ID):
    for B in DB:
        if B.booking_id == ID.upper():
            return B
    return None
def I(P, V=str):
    while True:
        try:
            U = input(P).strip()
            if not U:
                raise ValueError("Input cannot be empty.")
            return V(U)
        except ValueError as e:
            print(f"\n Invalid input: {e}. Please try again.")
def C():
    print("\n---  Create New Booking ---")
    try:
        N = I("Enter customer name: ")
        D = I("Enter destination (City, Country): ")
        T = I("Enter travel date (YYYY-MM-DD): ")
        R = I("Enter number of travelers: ", int)
        Y = I("Enter booking type (Flight/Hotel/Tour): ")
        B = Booking(N, D, T, R, Y)
        DB.append(B)
        print(f"\n SUCCESS: Booking created with ID: **{B.booking_id}**")
    except Exception as e:
        print(f"\n An unexpected error occurred during creation: {e}")
def A():
    if not DB:
        print("\n The booking list is currently empty.")
        return
    print("\n==================================")
    print("  **ALL TRAVEL BOOKINGS**")
    print("==================================")
    for B in DB:
        print(B)
    print(f"Total Bookings: {len(DB)}")
    print("==================================")
def S():
    ID = input("Enter the 8-character Booking ID to search: ").strip()
    B = F(ID)
    if B:
        print("\n **BOOKING FOUND:**")
        print(B)
    else:
        print(f"\n ERROR: No booking found with ID: **{ID.upper()}**")
def U():
    ID = input("Enter the 8-character Booking ID to update: ").strip()
    B = F(ID)
    if not B:
        print(f"\n ERROR: No booking found with ID: **{ID.upper()}**")
        return
    print(f"\n--- Updating Booking ID: **{B.booking_id}** ---")
    print(f"Current Destination: {B.destination}")
    ND = input("Enter new destination (leave blank to keep current): ").strip()
    if ND:
        B.destination = ND
    print(f"Current Status: {B.status}")
    NS = input("Enter new status (e.g., 'Cancelled', 'Completed', leave blank to keep current): ").strip()
    if NS:
        B.status = NS
    print(f"\n SUCCESS: Booking **{B.booking_id}** updated!")
    print(B)
def D():
    ID = input("Enter the 8-character Booking ID to delete: ").strip()
    B = F(ID)
    if not B:
        print(f"\n ERROR: No booking found with ID: **{ID.upper()}**")
        return
    print(f"\nBooking Found for deletion: {B.customer_name} to {B.destination}")
    C = input(f"Are you sure you want to delete booking **{B.booking_id}**? (y/n): ").lower().strip()
    if C == 'y':
        DB.remove(B)
        print(f"\n SUCCESS: Booking **{B.booking_id}** has been deleted.")
    else:
        print("\nAction cancelled.")
def M():
    print("\n\n---  Travel Booking System Menu ---")
    print("1. Create New Booking")
    print("2. Read / View All Bookings")
    print("3. Search Booking by ID")
    print("4. Update Booking Details")
    print("5. Delete Booking")
    print("6. Exit")
def main():
    DB.append(Booking("Alice Johnson", "London, UK", "2026-06-15", 2, "Flight"))
    DB.append(Booking("Bob Smith", "Bali, Indonesia", "2026-01-20", 4, "Hotel"))
    print(f"System loaded with **{len(DB)}** sample bookings.")
    while True:
        M()
        C = input("Enter your choice (1-6): ").strip()
        if C == '1':
            C()
        elif C == '2':
            A()
        elif C == '3':
            S()
        elif C == '4':
            U()
        elif C == '5':
            D()
        elif C == '6':
            print("\n Thank you for using the Travel Booking System. Goodbye!")
            sys.exit(0) 
        else:
            print("\n Invalid choice. Please enter a number from 1 to 6.")
if __name__ == "__main__":
    main()
```
